# Preprocess Data

## Grand Total & Customer Name

```python
missing_grand_total = df2[df2['Grand Total'].isna() & df2['Issued Date'].notna()]
missing_customer = df2[df2['Customer/Display Name'].isna() & df2['Issued Date'].notna()]

row_numbers = missing_grand_total.index.tolist()
row_numbers2 = missing_customer.index.tolist()

print(row_numbers)
print(row_numbers2)

# dapat kalau grand total krn pax 0 dan passenger false
# dapat kalau customer kosong, hanya ada di cancelled dan expired
```

## Birth Date

```python
# Lihat birth date anomali
df2['Passenger/Birth Date'] = pd.to_datetime(df2['Passenger/Birth Date'], errors='coerce')
birth_after_2024 = df2[df2['Passenger/Birth Date'].dt.year > 2024]

row_numbers = birth_after_2024.index.tolist()
print(row_numbers)

jumlah = (df2['Passenger/Birth Date'].dt.year > 2024).sum()
print(jumlah)
```

```python
# mengosongkan birth date yang anomali (>2024)
df2['Passenger/Birth Date'] = pd.to_datetime(df2['Passenger/Birth Date'], errors='coerce')
df2.loc[df2['Passenger/Birth Date'].dt.year > 2024, 'Passenger/Birth Date'] = pd.NaT
```

## Hapus Direction Other

```python
before_delete = len(df2)

rows_to_drop = []
i = 0
while i < len(df2):
    if df2.loc[i, 'Direction'] == 'Other':
        rows_to_drop.append(i)
        i += 1
        while i < len(df2) and (pd.isna(df2.loc[i, 'Direction']) or df2.loc[i, 'Direction'] == 'Other'):
            rows_to_drop.append(i)
            i += 1
    else:
        i += 1

df2 = df2.drop(index=rows_to_drop).reset_index(drop=True)

after_delete = len(df2)
deleted_rows = before_delete - after_delete
```

```python
# Tampilkan hasil
print("Total rows before deletion:", before_delete)
print("Total rows after deletion :", after_delete)
print("Deleted rows              :", deleted_rows)
```

```python
other_rows = (df2['Direction'] == 'Other').sum()
print("Rows with 'Other' in 'Direction':", other_rows)
```

## Hapus Baris Grand total & customer yang kosong

```python
before_delete = len(df2)

mask = (
    (df2['Grand Total'].isna() | df2['Customer/Display Name'].isna()) &
    df2['Issued Date'].notna()
)

rows_to_drop = df2[mask].index

df2 = df2.drop(index=rows_to_drop).reset_index(drop=True)

after_delete = len(df2)
deleted_rows = before_delete - after_delete

# Tampilkan hasil
print("Total rows before deletion:", before_delete)
print("Total rows after deletion :", after_delete)
print("Deleted rows              :", deleted_rows)
```

# Data untuk Visualisasi

## Total Pax

```python
df2['Journeys/Departure Date'] = pd.to_datetime(df2['Journeys/Departure Date'], errors='coerce')
df2['Journeys/Departure Date'] = df2['Journeys/Departure Date'].dt.date
df2['Bulan'] = pd.to_datetime(df2['Journeys/Departure Date']).dt.to_period('M')

pax_per_bulan = df2.groupby('Bulan')['Total Pax'].sum().reset_index()
pax_per_bulan['Bulan'] = pax_per_bulan['Bulan'].astype(str)
pax_per_bulan['Year'] = pd.to_datetime(pax_per_bulan['Bulan']).dt.year
pax_per_bulan['Month'] = pd.to_datetime(pax_per_bulan['Bulan']).dt.month_name()
```

## Rute (Negara)

```python
df2["Extracted"] = df2["Destination/Display Name"].str.extract(r"\((\b[A-Za-z]{3}\b)\)")

df_pax_sum = df2.groupby("Extracted")["Total Pax"].sum().reset_index()
df_pax_sum.columns = ["Extracted Name", "Total Pax Sum"]

file_path_airport = "airport-codes.csv"
df_airport = pd.read_csv(file_path_airport)

df_rute = df_pax_sum.merge(df_airport, left_on="Extracted Name", right_on="iata_code", how="left")

# Pisahkan koordinat menjadi latitude dan longitude
df_rute[['latitude', 'longitude']] = df_rute['coordinates'].str.split(',', expand=True)
df_rute['latitude'] = df_rute['latitude'].astype(float)
df_rute['longitude'] = df_rute['longitude'].astype(float)

# Dapatkan daftar unik negara (ISO Code) + opsi "All"
unique_countries = df_rute["iso_country"].dropna().unique()
country_options = [{"label": "Semua Negara", "value": "All"}] + [{"label": c, "value": c} for c in unique_countries]
```
